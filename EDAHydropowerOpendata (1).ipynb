```python
province_counts = df['Province'].value_counts().sort_index()
province_counts
```

```python
plt.figure(figsize=(10, 6))
plt.bar(province_counts.index, province_counts.values, color='skyblue')
plt.title('Number of Hydropower Projects per Province', fontsize=14)
plt.xlabel('Province', fontsize=12)
plt.ylabel('Number of Projects', fontsize=12)

plt.grid(axis='y', linestyle='--')

plt.show()
```

```python
'''we can see that the province 4 has highest number of projects, with above 175 and the province 5 has minimum number of projects. However, Provinces 5, 6,
and 7 have significantly fewer projects, each with around 25, indicating a more uneven distribution of hydropower development across the regions. 
while province 1 and province 3 have a significant number of projects of around 150.'''
```

```python
capacity_by_province = df.groupby('Province')['Capacity (MW)'].sum()
capacity_by_province
```

```python
plt.figure(figsize=(10, 6))
plt.bar(capacity_by_province.index, capacity_by_province.values, color='skyblue')
plt.title('Total Hydropower Capacity by Province (MW)', fontsize=14)
plt.xlabel('Province', fontsize=12)
plt.ylabel('Capacity of Hydropower Projects', fontsize=12)

plt.grid(axis='y', linestyle='--')

plt.show()
```

```python
'''we can obsserve that the province 4 has highest capacity among all the other proviences of around 6000MW. 
    The province with second highest capacity seems to be the province 1 with total capacity slightly higher than 5000MW followed by province 3 with 
    capacity slightly higher than 4000MW
    province 6 and province 7 have impressive capacity since we had observed that these provinces had as low as around 25 hydropower projects however the
    province 5 is seen to have minimum capacity when it also had the same number of projects as province 6 and province 7.'''
```

```python
df.rename(columns={'Isuue Date':'Issue Date'},inplace=True)
df
```

```python
df['Issue Date'] = pd.to_datetime(df['Issue Date'], format='%m/%d/%Y', errors='coerce')
df['Year'] = df['Issue Date'].dt.year
licenses_per_year = df['Year'].value_counts().sort_index()
print("Number of Licenses Issued Per Year:")
print(licenses_per_year)
```

```python
plt.figure(figsize=(12, 6))
licenses_per_year.plot(kind='line', marker='o', color='blue', linewidth=2)
plt.title('Evolution of Hydropower License Issuance Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Licenses Issued', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
'''we can see that the number of license issued between the year 2050 to 2066 remained consistently low, near zero 
    but we can observe gradual increase in the trendwith licenses rising to about 50 by 2070.
    The number of lincese issued significantly increased after the year 2070,sharp spike is observed 
    between 2070 and 2073, where issuance peaks at around 225 licenses However, this surge is followed by a steep decline to around 50 licenses by 2075.'''
```
